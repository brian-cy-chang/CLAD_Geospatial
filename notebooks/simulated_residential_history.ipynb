```python
import pandas as pd
import geopandas as gpd
import numpy as np
import os
```

```python
# locate the file types
shp_dir = os.path.join(os.getcwd(), '..', 'output/HIFLD/centroids')
file_types = [x for x in os.listdir(shp_dir) if not x.endswith('.csv')]
```

```python
# test find the shapefile name
[each for each in os.listdir(os.path.join(shp_dir, file_types[0])) if each.endswith('.shp')]
```

```python
file_types
```

```python
# random read 
for eachfiletype in file_types:
    # locate the shp file
    shp_src = os.path.join(shp_dir, 
                           eachfiletype, 
                           [each for each in os.listdir(os.path.join(shp_dir, eachfiletype)) if each.endswith('.shp')][0])
    if 'random_reads' not in locals():
        random_reads = gpd.read_file(shp_src)
    else:
        random_reads = random_reads.append(gpd.read_file(shp_src))

    print(eachfiletype, random_reads.shape[0])
    
print(random_reads.shape[0])
```

```python
# randomize order
tmp = random_reads.sample(random_state=100, n=20000)\
    .drop('Place_type', axis=1)

# assign user
np.random.seed(seed=10)
tmp['user_id'] = np.random.choice(np.arange(start=1, stop=5000, step=1), 
                                  20000, 
                                  replace=True)

# assign location start
dr = pd.date_range(start='2000-01-01', end='2024-01-01', freq='D')

np.random.seed(seed=11)
tmp['start_date'] = np.random.choice(dr, 20000, replace=True)

np.random.seed(seed=12)
tmp['end_duration']=np.random.choice(np.arange(365,1000), 20000, replace=True).astype('timedelta64[D]')
tmp['end_date'] = tmp.apply(lambda x: x.start_date+x.end_duration, axis=1)
```

```python
# quick print
tmp.sort_values(['user_id','start_date'])
```

```python
# retrieve filenames within target folder
obj1 = pd.Series(os.listdir(os.path.join(os.getcwd(), '../..')))

# filter to objects of name convention
itemlist = obj1.loc[obj1.apply(lambda x: x.startswith('test_address'))]

# identify the most recent version
max_item = itemlist.apply(lambda x: x.split('.')[0].split('v')[1]).astype(int).max()
```
